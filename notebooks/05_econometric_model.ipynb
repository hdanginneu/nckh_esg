### Equation
$$ Y_{i,t} = \alpha_i + \gamma_t + \beta Climate_{i,t} + \theta Controls_{i,t} + \epsilon_{i,t} $$

Where:
- $\alpha_i$: Bank Fixed Effects
- $\gamma_t$: Time Fixed Effects
